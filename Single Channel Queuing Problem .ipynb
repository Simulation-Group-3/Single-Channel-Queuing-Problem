```python
from scipy.stats import poisson

InterarrivalTime = poisson.rvs(mu=5.6, size=19)

print(InterarrivalTime)
```

```python
sum = 0
arrivalTime = [0]
for i in InterarrivalTime:
  sum += i
  arrivalTime.append(sum)
```

```python
print(arrivalTime)
```

```python
from scipy.stats import expon
ServiceTime = expon.rvs(scale=1,loc=0,size=20)

for i in ServiceTime:
  print(i)
```

```python
servicebegins = [0]
serviceends = []

for x in range(0, 20):
  if(x>0):
    servicebegins.append(max(arrivalTime[x], serviceends[x-1]))
    
  serviceends.append(ServiceTime[x] + servicebegins[x])
```

```python
print("Service Time begin")
for i in servicebegins:
  print(i)

print("----------------")
print("\nTime service ends\n")

for i in serviceends:
  print(i)
```

```python
waitingtime = [0]

for i in range(1,20):
  if(arrivalTime[i] < serviceends[i-1]):
    waitingtime.append(serviceends[i-1] - arrivalTime[i])
  else:
    waitingtime.append(0)

totalwaitingtime = 0
countcustomerwhowait = 0


for i in waitingtime:
  totalwaitingtime += i
  if(i>0):
    ++countcustomerwhowait
```

```python
for i in waitingtime:
  print(i)

print("\ntotal waiting time= ", totalwaitingtime)
print("total Customer who wait = ", countcustomerwhowait)
```

```python
customerspendSystem = []

for i in range(0, 20):
  customerspendSystem.append(ServiceTime[i] + waitingtime[i])
```

```python
for i in customerspendSystem:
  print(i)
```

```python
idletimeofservice = [0]

for i in range(1, 20):
  if(arrivalTime[i] > serviceends[i - 1]):
    idletimeofservice.append(arrivalTime[i] - serviceends[i - 1])
  else:
    idletimeofservice.append(0)
```

```python
for i in idletimeofservice:
  print(i)
```

```python
print("Table: Single Channel Queuing Problem\n")
print("Customer\t", "Interarrival Time\t", "Arrival Time\t", "Service Time\t", "Time service begins\t", "Waiting Time in Queue\t", "Time service ends\t", "Time Customer Spends in System\t", " \t  Idle Time of Server")
j = 0
for i in range(0, 20):
  print("  ",i+1 , '\t\t\t' ,InterarrivalTime[i-1] if i>0 else "-", "\t\t     ", arrivalTime[i], '\t   ', round(ServiceTime[i],2), '\t\t', round(servicebegins[i],2), '\t\t\t', round(waitingtime[i],2), '\t\t\t', round(serviceends[i],1), '\t\t\t\t', round(customerspendSystem[i],2), '\t\t\t\t', round(idletimeofservice[i],2))

```

```python
average_number_of_customer_waiting = countcustomerwhowait / 20
print("Average number of customer waiting = ", average_number_of_customer_waiting) 

average_waiting_time_for_a_customer = totalwaitingtime / 20
print("Average waiting time for a customer =", average_waiting_time_for_a_customer)
```
